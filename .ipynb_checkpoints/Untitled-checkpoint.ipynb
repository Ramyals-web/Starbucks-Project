```python
!pip install ipython-sql
import pandas as pd
import seaborn as sns 

import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sqlalchemy import create_engine
%load_ext sql

engine = create_engine('sqlite:///starbucks.db')
connection = engine.connect()
```

```python
connection.execute('drop table if exists starbucks_nutrition')
```

```python
df = pd.read_csv('starbucks_nutrition.csv').reset_index(drop=True)
df.head()
```

```python
df.to_sql('starbucks_nutrition', connection, if_exists='replace')
```

```python
%sql sqlite:///starbucks.db
```

```python
result = %sql select * from starbucks_nutrition
hd = result.DataFrame()
hd.head()
```

```python
filtered_keys = [k for k in hd.keys() if k not in ['index', 'Category', 'Beverage', 'Beverage_prep', 'Beverage_category']]
min_max_ranges = hd[filtered_keys].groupby('Size').agg(['min', 'max'])
```

```python
min_max_ranges.loc['Grande']
```

```python
average_sugar_all_drinks = round(hd[[' Sugars (g)','Size', 'Beverage_category']].groupby(['Beverage_category', 'Size']).mean())
grande_beverage=average_sugar_all_drinks.loc[(slice(None), 'Grande'), :]
grande_beverage
```

```python
# Rename columns
grande_renamed = grande_beverage.rename(columns={" Sugars (g)": "Sugar"})
grande_renamed
# grande_renamed.to_csv('grande_renamed_sugar.csv')
```

```python
clrs = ['grey' if x >= 120 else 'red' for x in grande_renamed['Sugar'] ]

# Create a bar chart based off of the group series from before
sugar_chart = grande_renamed.plot(kind='barh', color=clrs)

# Set the xlabel and ylabel using class methods
sugar_chart.set_xlabel("Sugar in grams")
sugar_chart.set_ylabel("Beverage Categories")
# sugar_chart.set_yticklabels(y_labels)
sugar_chart.set_title('Sugar content per Beverage Category')
sugar_chart.set_xlim(0, 60) # expand xlim to make labels easier to read
plt.figure(figsize=(12, 8))

plt.show()
plt.tight_layout()

```

```python
clrs = ['grey' if x <30 else 'red' for x in grande_renamed['Sugar'] ]
ax = grande_renamed.plot.barh(y="Sugar", rot=0, color=clrs)
    
# Show the plot
plt.show()
```
